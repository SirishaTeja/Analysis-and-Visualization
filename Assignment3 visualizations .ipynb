Matplotlib

```python
# plot numerical values
# to check progress 

# plt.plot -> line 
# it makes sense with timeseries datasets 

plt.plot(data['number_of_reviews']) # assumes x axis as range(0, data.shape[0])
```

```python
# Scatterplot of price vs. number_of_reviews
plt.scatter(data['price'],data['number_of_reviews'])

# Label the axes
plt.xlabel('Price')
plt.ylabel('Number of Reviews')

# Show the plot
plt.show()
```

```python

# Group the data by neighbourhood_group and calculate the mean price
grouped = data.groupby('neighbourhood_group')['price'].mean()

# Create a bar chart
grouped.plot(kind='bar')

# Set the title and labels
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')

# Show the plot
plt.show
```

```python
# Create a line chart of availability_365   vs. price
plt.plot(data['availability_365'], data['price'])
plt.xlabel('availability_365')
plt.ylabel('Price')
plt.title('Date vs. Price')
plt.show()
```

```python
# Read in the data
review_scores_rating = [90, 80, 70, 60, 50, 40, 30, 20, 10]

# Plot the histogram
plt.hist(review_scores_rating, bins=10, edgecolor='black')

# Add labels
plt.title('Review Scores Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python

import seaborn as sns

# Calculate the correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python

plt.scatter(data['neighbourhood_group'], data['price'])
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Scatterplot of Neighbourhood Group vs Price')
plt.show()
```

```python
# Create a list of continents
continents = data['neighbourhood_group'].unique()

# Create a list of prices by neighbourhood group
prices = data.groupby('neighbourhood_group')['price'].sum().tolist()

# Create a pie chart
plt.pie(prices, labels=continents, autopct='%1.1f%%')

# Show the chart
plt.show()
```
